```python
"""
* 
* 
"""
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

import plotly.graph_objs as go

# Create random data with numpy
import numpy as np

print(__version__) # requires version >= 1.9.0
init_notebook_mode(connected=True)


N = 1000
random_x = np.linspace(0, 1, N)
random_y0 = np.random.randn(N)+5
random_y1 = np.random.randn(N)
random_y2 = np.random.randn(N)-5

#print(random_y0)
random_y1[500]=50

# Create traces
trace0 = go.Scatter(
    x = random_x,
    y = random_y0,
    mode = 'lines',
    name = 'lines'
)
trace1 = go.Scatter(
    x = random_x,
    y = random_y1,
    mode = 'lines+markers',
    name = 'lines+markers'
)
trace2 = go.Scatter(
    x = random_x,
    y = random_y2,
    mode = 'markers',
    name = 'markers'
)
data = [trace0, trace1, trace2]
iplot(data, filename='line-mode')
```

```python
"""
 CSV file 
"""
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
import pandas as pd

print(__version__) # requires version >= 1.9.0
init_notebook_mode(connected=True)

# load data from csv file
df = pd.read_csv("D:/Python/dataset/for_plot_1.csv")
print(df.head())
#  dataframe  array
dfArray=df.values
print(len(dfArray))
#  array  list
x_list=[] # Initialize x_list as type list
y_list=[] # Initialize y_list as type list
for i in range(len(dfArray)):
    x_list.append(dfArray[i][0])
    y_list.append(dfArray[i][1])
print(x_list)
print(y_list)

# Create traces
trace0 = go.Scatter(
    x = x_list,
    y = y_list,
    mode = 'lines',
    name = 'lines'
)
data = [trace0]
iplot(data, filename='line-mode')
```
