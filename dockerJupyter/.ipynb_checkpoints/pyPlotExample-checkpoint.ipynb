```python
"""
Single line basic plot
"""
import matplotlib.pyplot as plt
import numpy as np

plt.subplot(111)
x_list=np.arange(0,360)
y_list=np.sin(x_list*np.pi/180.0)
plt.plot(x_list,y_list, label="sinewave")
# configure view window to use auto
plt.autoscale()
# adjust manually to zoom in/out
plt.xlim(-30,390)
plt.ylim(-1.5, 1.5)
# configure aspect (remark the following line ==> auto aspect)
plt.gca().set_aspect(aspect=100)
# labelling
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title( "The Title")
# display
plt.legend()
plt.show( )

# plt.savefig("/testPlot.png", dpi=300, format="png")
```

```python
"""
Multiple line basic plot
"""
import matplotlib. pyplot as plt
import numpy as np

plt.subplot(111)
x1=np.arange(0,360,5)
y1=np.sin(x1*np.pi/180.0)
x2=x1
y2=np.cos(x2*np.pi/180.0)
x3=x1
y3=np.tan(x3*np.pi/180.0)
plt.plot (x1,y1,lw=3, label='sin')
plt.plot(x2,y2,"ro", label='cosin')
plt.plot (x3,y3, "y--" ,lw=5, label='tan')
# configure view window to use auto
plt.autoscale()
# adjust manually to zoom in/out
#plt.xlim(-30,390)
plt.ylim(-5, 5)
# configure aspect (remark the following line ==> auto aspect)
#plt.gca().set_aspect(aspect=100)
# labelling
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title( "The Title")
# display
plt.legend()
plt.show()
```
